```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
# theta = np.zeros((n+1, 26))
# theta.size
```

```python
def model(X, Y, learning_rate, iterations):
    theta = np.zeros((n, 26))
    for i in range(iterations):
        Z = np.dot(X, theta)
        A = sigmoid(Z)
        
        cost = -((1/m)*np.sum( Y*np.log(A) + (1-Y)*np.log(1-A)))
        dtheta =  (1/m)*np.dot(X.T, (A-Y))
        theta = theta - learning_rate*dtheta
#         if(i%(iterations/10) == 0):
        print("The value of cost after", i, "iterations is", cost )
        plt.scatter(i,cost)   
    return theta,A
        
```

```python
iterations = 50
learning_rate = 0.1 
theta,A = model(X_train, Y, learning_rate, iterations)
```
